```python
from rectangle import Rectangle

# create a triangle
r1 = Rectangle(x=0, y=0, width=3, height=4)
print(r1)

# check area and perimeter
print(f"Area: {r1.area}")
print(f"Perimeter: {r1.perimeter}")

# check if it's a square
square = Rectangle(width=3, height=4)
print(f"Is square: {square.is_square()}")
```
